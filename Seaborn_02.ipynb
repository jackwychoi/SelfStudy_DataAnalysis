```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [3, 5, 8, 6, 7]
categories = ['A', 'B', 'A', 'B', 'A']
# Creating a scatter plot with color and hue using Seaborn
sns.scatterplot(x=x, y=y, hue=categories, palette='Set1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Color and Hue')
plt.show()
```

```python
# Sample data
data = sns.load_dataset('iris')
# Creating a basic box plot using Seaborn
sns.boxplot(x='species', y='sepal_length', data=data)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Basic Box Plot with Seaborn')
plt.show()
```

```python
# Sample data
data = sns.load_dataset('titanic')
# Creating a horizontal box plot with color palette using Seaborn
sns.boxplot(x='age', y='class', data=data, orient='h', palette='Set2')
plt.xlabel('Age')
plt.ylabel('Class')
plt.title('Horizontal Box Plot with Color Palette')
plt.show()
```

```python
# Sample data
data = sns.load_dataset('tips')
# Creating a grouped box plot using Seaborn
sns.boxplot(x='day', y='total_bill', data=data, hue='sex', palette='Set3')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Grouped Box Plot')
plt.legend(title='Gender')
plt.show()
```

```python
# Sample data
data = sns.load_dataset('diamonds')
# Creating a notched box plot using Seaborn
sns.boxplot(x='cut', y='price', data=data, notch=True, palette='pastel')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.title('Notched Box Plot')
plt.show()
```

```python
# Sample data
data = sns.load_dataset('mpg')
# Creating a customized box plot using Seaborn
sns.boxplot(x='origin', y='mpg', data=data, hue='cylinders', palette='Set2')
plt.xlabel('Origin')
plt.ylabel('Miles per Gallon')
plt.title('Customized Box Plot')
plt.legend(title='Cylinders')
plt.show()
```
