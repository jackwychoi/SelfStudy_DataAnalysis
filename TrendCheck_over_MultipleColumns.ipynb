```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(np.random.uniform(0,100000, size=(100, 26)), columns=list('ABCDEFGHIJKLMNOPQRSTUVWXYZ'))
```

```python
df.head()
```

```python
fig, axes = plt.subplots(4, 1, figsize=(14, 15), sharey=True)
# Plot the trends of the values over time for each of columns
for column in df.columns[:7]:
    sns.lineplot(ax=axes[0], x=df.index, y=df[column], label=column)

for column in df.columns[7:14]:
    sns.lineplot(ax=axes[1], x=df.index, y=df[column], label=column)
    
for column in df.columns[14:21]:
    sns.lineplot(ax=axes[2], x=df.index, y=df[column], label=column)
    
for column in df.columns[21:]:
    sns.lineplot(ax=axes[3], x=df.index, y=df[column], label=column)
    
axes[0].legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
axes[1].legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
axes[2].legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
axes[3].legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)

# Add a title and axis labels to the plot
axes[0].set_xlabel('Date')
axes[1].set_xlabel('Date')
axes[2].set_xlabel('Date')
axes[3].set_xlabel('Date')
axes[0].set_ylabel('')
axes[1].set_ylabel('')
axes[2].set_ylabel('')
axes[3].set_ylabel('')

# Display the plot
plt.show()
```
