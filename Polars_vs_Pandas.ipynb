```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
def polars():
    import polars as pl
    start_time = time.time()
    
    
    df = pl.DataFrame({'a': [1, 2, 3], 'b': [True, False, True]})

    df.write_csv('data/data_polars.csv')
    end_time = time.time()
    execution_time = end_time - start_time

    print(df)
    print(f'Execution time for Polars: {execution_time} seconds')

def pandas():
    import pandas as pd
    start_time = time.time()
    df = pd.DataFrame({'a': [1, 2, 3], 'b': [True, False, True]})

    df.to_csv('data/data_pandas.csv')

    end_time = time.time()

    execution_time = end_time - start_time

    print(df)
    print(f'Execution time for Pandas: {execution_time} seconds')
    
    
polars()
pandas()
```
