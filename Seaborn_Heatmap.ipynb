```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
df.corr()
```

```python
my_corr_mat = df.corr()
```

```python
sns.heatmap(my_corr_mat)
```

```python
sns.heatmap(my_corr_mat, cmap='crest')
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, cmap='plasma')
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, cmap='OrRd');
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, cmap='mako')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, cmap='mako', vmin=0.8, vmax=1);
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, cmap='PiYG', center=0.8, vmin=0.8, vmax=1);
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, annot=True);
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, annot=True, fmt='.3f');
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, annot=True, linewidths=3, linecolor='b', cmap='magma');
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, annot=True, cbar=False);
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, annot=True, square=True);
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(my_corr_mat, annot=True, mask=np.triu(my_corr_mat));
```
