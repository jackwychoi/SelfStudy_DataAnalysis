```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://github.com/shoukewei/data/raw/main/data-wpt/USD_CNY%20Historical%20Data.csv'
df = pd.read_csv(url)
df
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df[['ISO_Year', 'ISO_Week', 'ISO_Weekday']] = df['Date'].dt.isocalendar()
df
```

```python
df['ISO_Year'] = df['Date'].dt.isocalendar().year
df[['ISO_Year']]
```

```python
df['ISO_Week'] = df['Date'].dt.isocalendar().week
df[['ISO_Week']]
```

```python
df['ISO_Weekday'] = df['Date'].dt.isocalendar().day
df[['ISO_Weekday']]
```

```python
mondays = df[df['ISO_Weekday'] == 1]
mondays
```

```python
weekly_avg_exchange_rate = df.groupby('ISO_Week')['Price'].mean()
print(weekly_avg_exchange_rate)
```

```python
plt.plot(weekly_avg_exchange_rate)
plt.xlabel('ISO Week')
plt.ylabel('Average Exchange Rate')
plt.title('Average Exchange Rate per ISO Week')
plt.show()
```
