```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/college_major_women_stem.csv')
df.head()
```

```python
sns.relplot(x='ShareWomen', y = 'Median', data = df)
plt.show()
```

```python
sns.set_palette('Paired', 10)
sns.relplot(x='ShareWomen', y='Median', hue = 'Major_category',
             data = df)
plt.title('Median Salary vs Share of Women in a Particular Major')
plt.show()
```

```python
sns.relplot(x='ShareWomen', y='Median', hue = 'Major_category',
             style = 'Major_category', data = df)
plt.show()
```

```python
sns.relplot(x='ShareWomen', y='Median', hue = 'Total', data = df)
plt.title('Median Salary vs Share of Women in a Particular Major')
plt.show()
```

```python
sns.relplot(x='ShareWomen', y='Median', hue = 'Major_category', 
             size = 'Total', sizes = (15, 200), data = df)
plt.title('Median Salary vs Share of Women in a Particular Major')
plt.show()
```

```python
sns.relplot(x= 'ShareWomen', y = 'Women', kind = 'line', data = df)
plt.title('Share of Women vs Total Women in a Particular Major')
plt.show()
```

```python
sns.lmplot(x = 'ShareWomen', y='Median', data = df)
plt.show()
```

```python
sns.catplot(x = 'Major_category', y = 'Median', data = df)
plt.title('Median Salary of Different Major Categories')
plt.show()
```

```python
#Option 1
sns.catplot(x = 'Median', y = 'Major_category', data = df)
plt.title('Median Salary of Different Major Categories')
plt.show()
#Option 2
sns.catplot(x = 'Major_category', y = 'Median', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
df.sample(10)
```

```python
df['Gender_majority'] = None
df.loc[df['Men'] > df['Women'], 'Gender_majority'] = 'Majority Men'
df.loc[df['Women'] > df['Men'], 'Gender_majority'] = 'Majority Women'
```

```python
df['Gender_majority'].value_counts()
```

```python
sns.catplot(x = 'Major_category', y = 'Median', hue = 'Gender_majority', data = df)
plt.title('Median Salary of Different Major Categories')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Option 1
sns.catplot(x = 'Major_category', y = 'Median', jitter = False,
             data = df)
plt.xticks(rotation = 90)
plt.show()
# Option 2
sns.catplot(x = 'Major_category', y = 'Median', kind = 'swarm', 
            data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(x = 'Major_category', y = 'Median', kind = 'box', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Boxen Plot
sns.catplot(x = 'Major_category', y = 'Median', kind = 'boxen', 
             data = df)
plt.title('Boxen Plot')
plt.xticks(rotation = 90)
plt.show()
# Violin Plot
sns.catplot(x = 'Major_category', y = 'Median', kind = 'violin',   
            data = df)
plt.title('Violin Plot')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(x = 'Major_category', y = 'Median', kind = 'violin', 
            hue = 'Gender_majority', split = 'True', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(x = 'Major_category', kind = 'count', data = df)
plt.xticks(rotation = 90)
plt.show()
```
