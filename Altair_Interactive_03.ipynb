```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Pokemon.csv')
```

```python
df.head(2)
```

```python
alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color='Type 1'
).interactive()
```

```python
interval = alt.selection_interval()

alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color='Type 1'
).properties(
   selection=interval
)
```

```python
interval = alt.selection_interval()

alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color=alt.condition(interval, 'Type 1', alt.value('lightgray'))
).properties(
   selection=interval
)
```

```python
interval = alt.selection_interval(encodings=['y'])

alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color=alt.condition(interval, 'Type 1', alt.value('lightgray'))
).properties(
   selection=interval
)
```

```python
interval = alt.selection_interval(encodings=['y'])

scatter = alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color=alt.condition(interval, 'Type 1', alt.value('lightgray'))
).properties(
   selection=interval
)

bar = alt.Chart(df).mark_bar().encode(
   x='average(Total)', 
   y='Type 1', 
   color='Type 1'
).transform_filter(
   interval
)

# scatter | bar
# scatter & bar
# alt.hconcat(scatter, bar)
alt.vconcat(scatter, bar)
```

```python
multi = alt.selection_multi()

scatter = alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color=alt.condition(multi, 'Type 1', alt.value('darkgray'))
).properties(
   selection=multi,
   title='Total vs Speed'
)

bar = alt.Chart(df).mark_bar().encode(
   x='average(Total)', 
   y='Type 1', 
   color='Type 1'
)

scatter & bar
```

```python
multi = alt.selection_multi()

scatter = alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color=alt.condition(multi, 'Type 1', alt.value('darkgray'))
).properties(
   selection=multi,
   title='Total vs Speed'
)

bar = alt.Chart(df).mark_bar().encode(
   x='average(Total)', 
   y='Type 1', 
   color='Type 1'
)

scatter | bar
```

```python
multi = alt.selection_multi(fields=['Type 1'])
interval = alt.selection_interval(encodings=['x', 'y'], empty='all')

scatter = alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color=alt.condition(multi & interval, 'Type 1', alt.value('darkgray'))
).properties(
   selection=interval
)

bar = alt.Chart(df).mark_bar().encode(
   x='average(Total)', 
   y='Type 1', 
   color=alt.condition(multi, 'Type 1', alt.value('darkgray'))
).properties(
   selection=multi,
   title='Total'
).transform_filter(
   interval)

scatter | bar
```

```python
multi = alt.selection_multi(fields=['Type 1'])
interval = alt.selection_interval(encodings=['x', 'y'], empty='all')

scatter = alt.Chart(df).mark_point().encode(
   x='Total',
   y='Speed',
   color=alt.condition(multi & interval, 'Type 1', alt.value('darkgray'))
).properties(
   selection=interval
)

bar = alt.Chart(df).mark_bar().encode(
   x=alt.X('average(Total)', title='Total', 
           axis=alt.Axis(tickMinStep=100),
           scale=alt.Scale(domain=[0, 700])), 
   y=alt.Y('Type 1', title='Type', scale=alt.Scale(domain=df['Type 1'].sort_values(ascending=True).unique().tolist())),
   color=alt.condition(multi, 'Type 1', alt.value('darkgray'))
).properties(
   selection=multi,
   title='Total'
).transform_filter(
   interval)

scatter | bar
```
