```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {
    'Order Date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01'],
    'Country': ['USA', 'Canada', 'USA', 'Mexico', 'Canada'],
    'Sales': [1000, 1500, 800, 1200, 2000],
    'Quantity': [10, 12, 8, 15, 20],
    'Gross Sales': [1200, 1800, 1000, 1400, 2400],
    'Profit': [200, 300, 200, 200, 400],
    'Region': ['West', 'North', 'West', 'South', 'North']
}
sales_data = pd.DataFrame(data)
# Calculate Gross Margin
sales_data['Gross Margin'] = sales_data['Gross Sales'] - sales_data['Profit']
sales_data
```

```python
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data.set_index('Order Date', inplace=True)
sales_data.resample('M')['Gross Margin'].mean().plot(color='blue')
```

```python
sales_by_country = sales_data.groupby('Country')['Sales'].sum()
sales_by_country.plot(kind='bar', color='green')
```

```python
# Bar plot with vibrant colors
plt.figure(figsize=(8, 6))
sales_data.groupby('Country')['Sales'].sum().plot(kind='bar', colormap='viridis')
plt.title('Total Sales by Country')
plt.xlabel('Country')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_data.plot(kind='scatter', x='Quantity', y='Gross Sales', color='red')
```

```python
# Scatter plot with vibrant colors
plt.figure(figsize=(8, 6))
plt.scatter(sales_data['Quantity'], sales_data['Profit'], c=sales_data['Gross Sales'], cmap='inferno')
plt.title('Profit vs. Quantity with Gross Sales')
plt.xlabel('Quantity')
plt.ylabel('Profit')
plt.colorbar(label='Gross Sales')
plt.show()
```

```python
sales_data['Profit'].hist(color='orange')
```

```python
# Histogram with vibrant colors
plt.figure(figsize=(8, 6))
sales_data['Sales'].plot(kind='hist', bins=10, color='purple', edgecolor='white')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
sales_data.boxplot(column='Profit', by='Region', vert=False, color='purple')
```

```python
# Box plot with vibrant colors
plt.figure(figsize=(8, 6))
sales_data.boxplot(column='Profit', by='Region', color='green')
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.suptitle('')
plt.show()
```
