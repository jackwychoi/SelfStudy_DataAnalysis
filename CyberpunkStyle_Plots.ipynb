```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as font_manager
from mpl_toolkits.mplot3d import Axes3D
import mplcyberpunk
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import Font
fe = font_manager.FontEntry(
    fname='data/fonts/PTSerif-Italic.ttf',
    name='PTSerifItalic')
font_manager.fontManager.ttflist.insert(0, fe)
```

```python
# Define custom cyberpunk theme
cyberpunk_theme = {
    'axes.edgecolor': 'white',
    'axes.facecolor': '#1d1f21',
    'axes.labelcolor': 'white',
    'axes.titlecolor': 'white',
    'figure.facecolor': '#1d1f21',
    'xtick.color': 'white',
    'ytick.color': 'white',
    'text.color': 'white',
    'grid.color': 'gray',
    'grid.linestyle': ':'
}
```

```python
# Generate data for plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

print(x)
print(y)
```

```python
# Set cyberpunk theme as default
plt.style.use(cyberpunk_theme)

def neon_plot(x, y, ax=None):
    if ax is None:
        ax = plt.gca()
    line, = ax.plot(x, y, lw=1, zorder=6)
    for cont in range(6, 1, -1):
        ax.plot(x, y, lw=cont, color=line.get_color(), zorder=5,
                alpha=0.05)
    return ax

# Create plot
fig, ax = plt.subplots(figsize=(12, 8))
neon_plot(x, y)
ax.set(xlabel='Time (s)', ylabel='Amplitude')
ax.set_title('Sine plot in Cyberpunk style', font='PTSerifItalic', fontsize=25)

# Add neon highlights
ax.spines['left'].set_color('#66ccff')
ax.spines['bottom'].set_color('#66ccff')
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.tick_params(axis='x', colors='#66ccff')
ax.tick_params(axis='y', colors='#66ccff')
ax.grid(True)

plt.show()
```

```python
# Generate random data for 3D surface plot
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)
r = np.sqrt(x ** 2 + y ** 2)
z = np.sin(r)

print(x, y)
print(r)
print(z)
```

```python
# Create 3D surface plot with neon green highlights
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='Greens')
ax.set(xlabel='X', ylabel='Y', zlabel='Z')
ax.set_title('Three D Surface plot in Cuberpunk style', font='PTSerifItalic', fontsize=25)
ax.tick_params(axis='x', colors='#66ccff')
ax.tick_params(axis='y', colors='#66ccff')
ax.tick_params(axis='z', colors='#66ccff')
ax.xaxis.line.set_color('#66ccff')
ax.yaxis.line.set_color('#66ccff')
ax.zaxis.line.set_color('#66ccff')
ax.xaxis.label.set_color('#66ccff')
ax.yaxis.label.set_color('#66ccff')
ax.zaxis.label.set_color('#66ccff')
ax.grid(True)

plt.show()
```
