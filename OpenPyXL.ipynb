```python
import pandas as pd
from openpyxl import Workbook
```

```python
ps_games = pd.read_csv('data/playstation_4_games.csv')
ps_sales = pd.read_csv('data/PS4_GamesSales.csv', header=0, encoding='unicode_escape')
all_sales = pd.read_csv('data/Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
# Create an Excel file
save = pd.ExcelWriter('data/Games_Datasets.xlsx', engine='xlsxwriter')

# Create tabs
all_sales.to_excel(save, sheet_name='All Sales', index=False)
ps_games.to_excel(save, sheet_name='PS Games', index=False)
ps_sales.to_excel(save, sheet_name='PS Games Sales', index=False)

# Save and close Excel
save.close()
```
