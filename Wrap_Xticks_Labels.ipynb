```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['neighborhood', 'accommodates', 'price']
airbnb = pd.read_csv('data/airbnb.csv', usecols=cols)
airbnb.head()
```

```python
airbnb['neighborhood'].replace({1:'Capitol Hill', 2:'Union Station', 3:'Columbia Heights',
                                4:'Dupont Circle', 5:'Shaw', 6:'Edgewood', 7:'Brightwood Park',
                                8:'Downtown', 9:'Kalorama Heights', 10:'Howard University'}, inplace=True)
```

```python
sns.set_theme(style='darkgrid', rc={'figure.dpi': 147},              
              font_scale=0.7)
fig, ax = plt.subplots(figsize=(7, 2))
top10 = airbnb['neighborhood'].value_counts().index[:10]
ax.set_title('Mean Price per Neighborhood')
sns.barplot(x='neighborhood', y='price', data=airbnb, ax=ax, 
            order=top10, ci=None, 
            hue='accommodates', hue_order=[1, 2, 3])
ax.legend(bbox_to_anchor=(1, 1), title='accommodates');
```

```python
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.figure
```

```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)
```

```python
wrap_labels(ax, 10)
ax.figure
```
