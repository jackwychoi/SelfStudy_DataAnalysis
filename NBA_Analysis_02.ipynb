```python
import pandas as pd
from nba_api.stats.endpoints import TeamGameLogs
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_seq_items', None)
```

```python
# List of season years
season_years = ['2023-24', '2022-23', '2021-22', '2020-21', '2019-20', '2018-19']
```

```python
# Create an empty DataFrame to store the data
games = pd.DataFrame()

# Loop through the season years and append to the games df
for season_year in season_years:

    gamedatapull = TeamGameLogs(
        league_id_nullable='', #default is the nba, can adj to g-league,wnba
        team_id_nullable='', # will pull all teams from the league chosen, can adj to select specfic team_id
        season_nullable = season_year, # inputting our variable 
        season_type_nullable='Regular Season' # can incldue Pre Season, Playoffs, All Star
    )
    
    games_season = gamedatapull.get_data_frames()[0]
    games = pd.concat([games, games_season], ignore_index=True)
```

```python
games.head(3)
```

```python
# Convert the 'date' column to datetime format
games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])

# Convert the datetime format to your desired format
games['GAME_DATE'] = games['GAME_DATE'].dt.strftime('%m/%d/%Y')
```

```python
games.head(3)
```

```python
# drop "Rank" columns since they seem to be unaccurate. 
# note that columns must be contigous or next to each other for code to work
games.drop(columns=games.loc[:,'GP_RANK':'PLUS_MINUS_RANK'], inplace=True)
```

```python
games.head(3)
```

```python
games_pv = games.loc[games['WL']=='W'].pivot_table(index='TEAM_NAME', columns='SEASON_YEAR', values=['WL'], aggfunc='count')
```

```python
games_pv.head(3)
```

```python
games_pv['Total Wins'] = games_pv[games_pv.columns].sum(axis=1)
games_pv.sort_values('Total Wins', ascending=False, inplace=True)
games_pv.rename(columns=lambda c: c.replace('WL', 'Wins'), inplace=True)
```
