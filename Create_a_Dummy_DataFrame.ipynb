```python
import pandas as pd
import numpy as np
```

```python
# Generating dates between '2024-02-01' and '2024-02-15'
dates = pd.date_range(start='2024-02-01', end='2024-02-15', freq='D')

# Generating NE_IDs
ne_ids = np.random.randint(1000, 9999, size=10)

# Initialize lists to store data
data = {'NE_ID': [], 'gNB_ID': [], 'CellIdentity': []}

# Generating data
for i, ne_id in enumerate(ne_ids):
    # Randomly choosing the number of gNB_IDs for each NE_ID
    num_gnb_ids = np.random.randint(2, 5)
    gnb_ids = np.random.choice(np.arange(100000, 999999), size=num_gnb_ids, replace=False)

    for gnb_id in gnb_ids:
        # Randomly choosing the number of CellIdentities for each gNB_ID
        num_cell_ids = np.random.randint(3, 21)
        cell_ids = np.random.randint(0, 2001, size=num_cell_ids)

        # Appending data to lists for each date
        for cell_id in cell_ids:
            #print(ne_id, gnb_id, cell_id)
            data['NE_ID'].extend([ne_id])
            data['gNB_ID'].extend([gnb_id])
            data['CellIdentity'].extend([cell_id])

df = pd.DataFrame(data)
df.sort_values(['NE_ID', 'gNB_ID', 'CellIdentity'], ascending=True, inplace=True)
df['NE_gNB_Cell_ID'] = df['NE_ID'].astype(str)+'_'+df['gNB_ID'].astype(str)+'_'+df['CellIdentity'].astype(str)
df.drop_duplicates('NE_gNB_Cell_ID', inplace=True)

df_list = []
for i, date in enumerate(dates):
    locals()['df'+str(i)] = df.copy()
    locals()['df'+str(i)]['Date'] = date
    df_list.append(locals()['df'+str(i)])

df = pd.concat(df_list, ignore_index=True)
df = df[['Date', 'NE_ID', 'gNB_ID', 'NE_gNB_Cell_ID']]

print(df.shape)
print(df['NE_gNB_Cell_ID'].nunique()*df['Date'].nunique())
```
