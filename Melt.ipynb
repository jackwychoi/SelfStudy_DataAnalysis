```python
import pandas as pd
```

```python
df = pd.DataFrame({'Target': [1234, 1222, 2513, 2236, 3155],
                   'Board_ID_a': [0, 2, 1, 1, 0],
                   'Port_ID_a': [7, 1, 2, 10, 0],
                   'Unit_ID_a': [0, 0, 0, 0, 0],
                   'Switch_a': ['on', 'on', 'on', 'on', 'on'],
                   'Board_ID_b': [0, 2, 1, 1, 0],
                   'Port_ID_b': [8, 3, 2, 11, 1],
                   'Unit_ID_b': [1, 0, 0, 1, 0],
                   'Switch_b': ['on', 'on', 'on', 'on', 'on'],
                   'Board_ID_c': [1, 2, 1, 1, 0],
                   'Port_ID_c': [7, 4, 4, 12, 2],
                   'Unit_ID_c': [0, 0, 0, 0, 2],
                   'Switch_c': ['on', 'on', 'on', 'on', 'on']
                   })

df
```

```python
df['A'] = df['Board_ID_a'].astype(str)+'_'+df['Port_ID_a'].astype(str)+'_'+df['Unit_ID_a'].astype(str)+'_'+df['Switch_a'].astype(str)
df['B'] = df['Board_ID_b'].astype(str)+'_'+df['Port_ID_b'].astype(str)+'_'+df['Unit_ID_b'].astype(str)+'_'+df['Switch_b'].astype(str)
df['C'] = df['Board_ID_c'].astype(str)+'_'+df['Port_ID_c'].astype(str)+'_'+df['Unit_ID_c'].astype(str)+'_'+df['Switch_c'].astype(str)
```

```python
df = df[['Target', 'A', 'B', 'C']]
```

```python
df
```

```python
melted = df.melt(id_vars='Target', var_name='Variable', value_name='Value')
```

```python
melted.head(3)
```

```python
melted.tail(3)
```

```python
melted = melted.sort_values(['Target', 'Variable'], ascending=True)
```

```python
melted.head()
```

```python
melted['Board_ID'] = melted['Value'].str.split('_', expand=True)[0]
melted['Port_ID'] = melted['Value'].str.split('_', expand=True)[1]
melted['Unit_ID'] = melted['Value'].str.split('_', expand=True)[2]
melted['Switch'] = melted['Value'].str.split('_', expand=True)[3]
```

```python
melted.head()
```

```python
final = melted[['Target', 'Board_ID', 'Port_ID', 'Unit_ID', 'Switch']]
```

```python
final.reset_index(drop=True, inplace=True)
```
