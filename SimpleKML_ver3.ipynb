```python
import pandas as pd
import simplekml
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('data/korea_gmonpro_20230103_152811_Sim1_45006.csv')
```

```python
df_386 = df.loc[df['PCI/PSC/BSIC']==386]
df_151 = df.loc[df['PCI/PSC/BSIC']==151]
```

```python
def create_kml(df, filename, custom_color):
    kml = simplekml.Kml()
    for index, row in df.iterrows():
        pnt = kml.newpoint(name=row['xNBID'], coords=[(row['LON'], row['LAT'])])
        pnt.iconstyle.icon.href = 'antenna.png'
        pnt.labelstyle.color = simplekml.Color.rgb(0, 0, 0, 0)
        pnt.style.iconstyle.color = custom_color
        for col in df.columns:
            pnt.extendeddata.newdata(col, row[col])
    kml_filename = 'data/' + filename + '.kml'
    kml.save(kml_filename)
    
color_green = simplekml.Color.green
color_blue = simplekml.Color.blue
```

```python
create_kml(df_386, 'df_386', color_green)
```

```python
create_kml(df_151, 'df_151', color_blue)
```
