```python
import pandas as pd 
import numpy as np
```

```python
df = pd.DataFrame(
    np.abs(np.random.randn(10, 3)),   
    columns=list("ABC"),
)

df.iloc[2,0]=-1
df.iloc[5,1]=-1
df.iloc[2,1]=999
df.iloc[0,0]=np.nan
df.iloc[6,2]=None
```

```python
df
```

```python
df = df.replace([-1,999, None], np.nan)
df
```

```python
print(df.info())
print(df.isna().sum())
```

```python
df.dropna(subset=['A','B'], how='all')
```

```python
# Fill with default value
df.fillna(value=0)
# Forward fill
df.ffill()
# Fill with mean
df.fillna(value=df.mean())
```

```python
df.interpolate(method ='linear', limit_direction ='both')
```
