```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_seq_items', None)
pd.set_option('display.max_rows', None)
```

```python
data = {'product_name': ['pencil', 'eraser', 'pen', 'desk', 'laptop','laptop','pen','pencil','pen','eraser','gift','pencil',np.nan],
        'price': [2, 1, 1, 50, 678,500,2,5,3,-9,-2,-1,10], 
        'owner': ['Chi','Joe','Dan','Lisa','Rose','Chi','Dan','Lisa','Joe','Rose','Jen','Dan','Mike']}

df = pd.DataFrame(data)
df.head()
```

```python
df.style.hide_columns('price')
```

```python
df.style.highlight_null(null_color='yellow')
```

```python
df.groupby('owner').agg({'price':'sum'}).style.highlight_max(color='yellowgreen') 
```

```python
df.groupby('owner').agg({'price':'sum'}).style.highlight_min(color='salmon')
```

```python
df[['price']].style.highlight_quantile(q_left=0.9, axis=None, color='gold') 
```

```python
np.random.seed(100)
df2 = pd.DataFrame({'A': np.linspace(1, 8, 10)})
df2 = pd.concat([df2, pd.DataFrame(np.random.randn(8, 4), columns=list('BCDE'))],
               axis=1)
df2.iloc[3, 1] = np.nan
df2.iloc[1, 1] = np.nan
df2.head()
```

```python
cm = sns.light_palette('purple', as_cmap=True)
df2.style.background_gradient(cmap=cm) 
```

```python
cm = sns.light_palette('red', as_cmap=True)
df2.style.text_gradient(cmap=cm)
```

```python
df2.style.bar(subset=['A', 'B'], align='zero', color=['crimson', 'yellowgreen'])
```

```python
df2.style.set_caption('Data').background_gradient()
```

```python
def mean_greater(m):
    greater = m > m.mean()
    return ['background-color: dodgerblue' if i else 'background-color: hotpink' for i in greater]

df2.style.apply(mean_greater)
```
