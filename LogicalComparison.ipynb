```python
import pandas as pd
import yfinance as yf
```

```python
# fixed data so sample data will stay the same
df = yf.download('MSFT')
df = df.tail(10)
df
```

```python
old = df['Open'] >= 230
old
```

```python
df = df.astype({'Open':'float',
                'High':'float',
                'Low':'float',
                'Close':'float',
                'Adj Close':'float',
                'Volume':'float'})
```

```python
df.dtypes
```

```python
# is the adj close different from the close?
df['Close Comparison'] = df['Adj Close'].ne(df['Close'])
df
```

```python
# is the close greater than the open?
df['Bool Price Increase'] = df['Close'].gt(df['Open'])
df
```

```python
# was the volume greater than 20m?
df['High Volume'] = df['Volume'].ge(20000000)
df
```

```python
# was the close greater than yesterday's close?
df['Close (t-1)'] = df['Close'].shift(-1)
df['Bool Over Time Increase'] = df['Close'].gt(df['Close'].shift(-1))
df
```

```python
# did the open and close price match the predictions?
predictions = [236.96, 223.36, 241.01, 239.82, 221, 245, 202.84, 224.93, 310, 193.93]
df2 = df[['Open','Close']].round(2).eq(predictions, axis='index')
df2
```
