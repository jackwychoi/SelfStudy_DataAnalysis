```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)   #Create a 2D grid
Z = f(X, Y)
```

```python
plt.style.use('grayscale')
plt.contour(X, Y, Z, colors='black')
plt.show()
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy')  #20 contour lines
plt.show()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()    #Create a color bar to show data range of colors
plt.show()
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)       #Label the contour lines

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',  #Transfer grid visuals to an image
           cmap='RdGy', alpha=0.5)                  #alpha 0.5 means half-transparent
plt.colorbar()
plt.show()
```
