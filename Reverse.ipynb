```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/NourozR/Stock-Price-Prediction-LSTM/master/apple_share_price.csv'
df = pd.read_csv(url)
df
```

```python
df_reversed = df[::-1]
df_reversed
```

```python
df_reversed = df[::-1].reset_index(drop=True)
df_reversed
```

```python
df_reversed = df.reindex(index=df.index[::-1]).reset_index(drop=True)
df_reversed
```

```python
df_reversed = df.iloc[::-1].reset_index(drop=True)
df_reversed
```

```python
df_reversed = df.sort_index(ascending=False).reset_index(drop=True)
df_reversed
```

```python
df_reversed = df.sort_index(ascending=False,ignore_index=True)
df_reversed
```

```python
df.insert(1, 'Day', df['Date'].str.split('-', expand=True)[0])
df.insert(1, 'Month', df['Date'].str.split('-', expand=True)[1])
df.insert(1, 'Year', df['Date'].str.split('-', expand=True)[2])
df
```

```python
df['Day'].replace({'1':'01', '2':'02', '3':'03', '4':'04', '5':'05', 
                    '6':'06', '7':'07', '8':'08', '9':'09'}, regex=False, inplace=True)
df['Day'].unique()
```

```python
df['Month'].replace({'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06',
                    'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}, regex=False, inplace=True)
df
```

```python
df['Year'].unique()
```

```python
df['Year'] = '20'+df['Year']
df['Year'].unique()
```

```python
df['Date'] = df['Year']+'-'+df['Month']+'-'+df['Day']
print(df['Date'].dtype)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%Y-%m-%d')
print(df['Date'].dtype)
df.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)
df_reversed = df.sort_values(by=['Date'], ascending=True).reset_index(drop=True)
df_reversed
```
