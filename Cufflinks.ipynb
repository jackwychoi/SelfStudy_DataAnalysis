```python
import pandas as pd
import cufflinks as cf
import numpy as np
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
cf.set_config_file(theme='henanigans', offline=True)
# ggplot, pearl, solar, space, white, polar, henanigans
```

```python
cf.datagen.lines(4,1000).iplot()
```

```python
cf.datagen.sinwave(10,.25).iplot(kind='surface')
```

```python
cf.datagen.bubble(prefix="planet").iplot(kind='bubble', x='x', y='y', size='size', categories='categories', text='text',  
                                         xTitle='Relative Mass', yTitle='Distance from the Sun', title='Plannet Size, Mass and Distance')
```

```python
cf.datagen.heatmap(20,20).iplot(kind='heatmap', colorscale='spectral')
```

```python
cf.datagen.lines(3).iplot(kind='spread', xTitle='Dates', yTitle='Return')
```
