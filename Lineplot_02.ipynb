## Imports

```python
import matplotlib.pyplot as plt
import numpy as np

# Optional imports
from matplotlib import rcParams
from matplotlib import font_manager
```

## Settings

```python
for font in font_manager.findSystemFonts(fontpaths=None, fontext='ttf'):
    font_manager.fontManager.addfont(font)
```

```python
rcParams.update(
    {
        "font.family": "Roboto",
        "figure.figsize": (20, 9),
        "figure.facecolor": "#fbfbfb",
        "figure.titlesize": 40,
        "figure.titleweight": "bold",
        "font.size": 25,
        "lines.linewidth": 3,
        "grid.linewidth": 1.25,
        "grid.color": "#f6f5f4",
        "xtick.bottom": False,
        "ytick.left": False,
        "axes.spines.right": False,
        "axes.spines.top": False,
        "axes.edgecolor": "#c0bfbc"  # spine colour
    }
)
```

## Plot template

```python
fig, axs = plt.subplots(
    1, 3,
    sharex=True,
    sharey=True,
    layout="constrained"
)
fig.suptitle("The TITLE")

# Axis limits
axs[0].set_xlim(0, 10)
axs[0].set_ylim(0, 10)

# PLOTS

# Plot functions go here
# ...

# Other settings
for ax in axs:
    ax.set_xticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{x}" for x in np.arange(2, 10, 2)] + [""],
    )
    ax.set_yticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{y}" for y in np.arange(2, 10, 2)] + [""],
    )
    ax.grid()
```

## Axes lines

```python
# Example A
axs[0].set_title("A")
axs[0].axhline(
    y=5  # data coordinates
)
```

```python
# Example B
axs[1].set_title("B")
axs[1].axhline(
    y=3,  # data coordinates
    xmax=.5  # axe coordinates [0, 1]
)
axs[1].axhline(
    y=7,  # data coordinates
    xmin=.5  # axe coordinates [0, 1]
)
```

```python
# Example C
axs[2].set_title("C")
axs[2].axhline(
    y=1,
    xmin=.2,  # axe coordinates [0, 1]
    xmax=.8  # axe coordinates [0, 1]
)
axs[2].axhline(
    y=5,
    xmin=.3,  # axe coordinates [0, 1]
    xmax=.7  # axe coordinates [0, 1]
)
axs[2].axhline(
    y=9,
    xmin=.4,  # axe coordinates [0, 1]
    xmax=.6  # axe coordinates [0, 1]

)
```

```python
# Example D
axs[0].set_title("D")
axs[0].axvline(
    x=5
)
```
