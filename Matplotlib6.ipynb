```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generating sample data
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
sales = np.array([10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000])

# Plotting the data
plt.plot(months, sales)

# Customizing the axis labels and title
plt.xlabel('Month')
plt.ylabel('Sales in Dollars')
plt.title('Sales Trend over the Last Year')

# Displaying the plot
plt.show()
```

```python
# Generating sample data
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
temperature_f = np.array([40, 45, 55, 65, 75, 85, 95, 90, 80, 70, 55, 45])

# Convert Fahrenheit to Celsius
temperature_c = (temperature_f - 32) * 5 / 9

# Plotting the data
plt.plot(months, temperature_c)

# Customizing the ticks and tick labels
plt.xticks(np.arange(len(months)), months)
plt.yticks(np.arange(0, 50, 5))
plt.tick_params(axis='both', labelsize=12, pad=8, width=2)

# Formatting the tick labels
plt.gca().yaxis.set_major_formatter('{x:.1f} C')

# Adding grid lines
plt.grid(True)

# Adding axis labels and title
plt.xlabel('Month', fontsize=14, fontweight='bold')
plt.ylabel('Temperature (C)', fontsize=14, fontweight='bold')
plt.title('Temperature Trend over the Last Year', fontsize=16, fontweight='bold')

# Displaying the plot
plt.show()
```

```python
# Generating sample data
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
product_a_sales = np.array([1000, 1100, 1300, 1500, 1700, 1800, 2000, 1900, 1700, 1500, 1300, 1100])
product_b_sales = np.array([800, 900, 1000, 1200, 1400, 1500, 1600, 1500, 1300, 1100, 900, 800])

# Plotting the data
plt.plot(months, product_a_sales, label='Product A')
plt.plot(months, product_b_sales, label='Product B')

# Customizing the legend
plt.legend(loc='upper left', fontsize=12, framealpha=0.5, title='Product Sales', title_fontsize=14)

# Adding grid lines
plt.grid(True)

# Adding axis labels and title
plt.xlabel('Month', fontsize=14, fontweight='bold')
plt.ylabel('Sales (USD)', fontsize=14, fontweight='bold')
plt.title('Product Sales Trend over the Last Year', fontsize=16, fontweight='bold')

# Displaying the plot
plt.show()
```

```python
# Generating sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating the plot
plt.plot(x, y1, 'r-', label='sin(x)')
plt.plot(x, y2, 'b--', label='cos(x)')

# Adding grid lines
plt.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.5)

# Adding legend
plt.legend()

# Adding axis labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sin(x) and Cos(x)')

# Displaying the plot
plt.show()
```

```python
# Generating sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# Creating the plot with a red line
plt.plot(x, y, 'r-')

# Creating the plot with a dashed line
plt.plot(x, -y, 'r--')

# Creating the plot with a blue line
plt.plot(x, z, 'b-')

# Creating the plot with a custom color
plt.plot(x, -z, color=(0.2, 0.5, 0.1))

# Displaying the plot
plt.show()
```

```python
# Generate some random data
x = np.linspace(0, 10, 100)
y = np.sin(x)
# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, label='Sin', color='blue')
# Add text to the plot
ax.text(5, 0.8, 'Sin Wave', fontsize=12, ha='center', va='center')
# Customize the axis labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.set_title('Sin Wave with Annotations')
# Customize the legend
ax.legend(loc='upper right', fontsize=10)
# Show the plot
plt.show()
```

```python
# Generate some random data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y, label='Sin', color='blue')

# Add text to the plot
ax.text(5, 0.8, 'Sin Wave', fontsize=12, ha='center', va='center')
ax.annotate('Maximum', xy=(np.pi/2, 1), xytext=(3, 1.5), 
            arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

# Customize the axis labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.set_title('Sin Wave with Annotations')

# Customize the legend
ax.legend(loc='upper right', fontsize=10)

# Show the plot
plt.show()
```
