```python
from unprint import unprint

print('apple')
print('orange')
print('pear')

unprint(2) # this unprints 2 lines

print('pineapple')
```

```python
print('apple')
print('orange')
print('pear')

unprint(2, delay=1)
# this unprints 2 lines after a delay of 1 second

print('pineapple')
```

```python
from time import sleep

def progress_bar(n):
    print(f'[{'#'*n}{'-'*(20-n)}] {n/0.2}%')

for i in range(1, 21):
    unprint()
    progress_bar(i)
    sleep(0.2)
```

```python
a = r'''
    _\_/          
   /  / \ _ _ _ _ _
   \_/  /_ _ _ _ \  \ _ _
       / /      \ \   
      / /        \ \
'''

b = r'''
    _\_/              _ _
   /  / \ _ _ _ _ _ /
   \_/  /_ _ _ _ \
        \ \    / /   
         \ \  / /
'''

d = {a:b, b:a}
current = a
while True:
    print(current)
    unprint(7, delay=0.2)
    current = d[current]
```
