```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(0, 100000, size=(4, 8)), 
                  index=pd.date_range('2016-01-01', periods=4),
                  columns=pd.MultiIndex.from_tuples([('A', 'ENDC'), ('A', 'NRDC'), ('B', 'ENDC'), ('B', 'NRDC'), ('C', 'ENDC'), ('C', 'NRDC'), ('D', 'ENDC'), ('D', 'NRDC')]))
df
```

```python
# Calculate the sum of values per 'ENDC' column
sum_per_endc = df.xs('ENDC', level=1, axis=1).sum()
sum_per_endc
```

```python
# Sort the DataFrame by the sum of 'ENDC' values in descending order
endc_sorted_df = df[sum_per_endc.sort_values(ascending=False).index].copy()
endc_sorted_df
```

```python
# Calculate the sum for the last row and assign it to 'Total'
endc_sorted_df.loc['Total'] = endc_sorted_df.sum()
endc_sorted_df
```

```python
# Calculate the sum of values per 'NRDC' column
sum_per_nrdc = df.xs('NRDC', level=1, axis=1).sum()
sum_per_nrdc
```

```python
# Sort the DataFrame by the sum of 'NRDC' values in descending order
nrdc_sorted_df = df[sum_per_nrdc.sort_values(ascending=False).index].copy()
nrdc_sorted_df
```

```python
# Calculate the sum for the last row and assign it to 'Total'
nrdc_sorted_df.loc['Total'] = nrdc_sorted_df.sum()
nrdc_sorted_df
```
