```python
fig=plt.figure(figsize=(8,8))
plt.rcParams['figure.facecolor'] = 'white'
```

```python
label=['Poor', 'Fair', 'Good',' Excellent']
# The numbers are the portion of area, the summary of them are 100
val=[55,   # 100-45
    20,    # 45-25
    15,    # 25-10
    10]    # 10-0
```

```python
label.append('')
val.append(100)
```

```python
def donut(dr): # dr is an integer but means a % rate
    colors=['#FF0000', '#FFC000', '#FFFF00', '#92D050'] # Colors for each group

    wedges, labels=plt.pie(val, labels=label, colors=colors, wedgeprops=dict(width=0.4, edgecolor='w'))
    plt.title('Deficiency Rate Indicator', fontdict={'fontsize': 24})

    angle=math.radians(180 * dr / 100) # The angle of the arrow pointing at, in radian unit
    plt.annotate(dr,  
                 xy=(math.cos(angle) * -1, math.sin(angle)),       # Arrow sharp position
                xytext = (-0.2, 0),                                # Metric number position
                arrowprops=dict(facecolor='black', shrink=0.05),   # Arrow style
                fontsize=24
                )
#     wedges[-1].set_visible(False)                                  
    plt.show()
```

```python
donut(30)
```

```python
def half_donut(dr): # dr is an integer but means a % rate
    colors=['#FF0000', '#FFC000', '#FFFF00', '#92D050'] # Colors for each group

    wedges, labels=plt.pie(val, labels=label, colors=colors, wedgeprops=dict(width=0.4, edgecolor='w'))
    plt.title('Deficiency Rate Indicator', fontdict={'fontsize': 24})

    angle=math.radians(180 * dr / 100) # The angle of the arrow pointing at, in radian unit
    plt.annotate(dr,  
                 xy=(math.cos(angle) * -1, math.sin(angle)),       # Arrow sharp position
                xytext = (-0.2, 0),                                # Metric number position
                arrowprops=dict(facecolor='black', shrink=0.05),   # Arrow style
                fontsize=24
                )
    wedges[-1].set_visible(False)                                  
    plt.show()
```

```python
half_donut(30)
```

```python
half_donut(7)
```

```python
half_donut(23)
```
