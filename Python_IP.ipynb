```python
import ipaddress
from pprint import pprint
```

```python
x = list(ipaddress.ip_network('192.168.0.0/30', False))

# there are 4 ip addresses in this range
pprint(x)
```

```python
x = list(ipaddress.ip_network('192.168.0.0/28', False))

# there are 16 ip addresses in this range
pprint(x)
```
