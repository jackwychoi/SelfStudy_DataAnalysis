```python
# create dataset 1
x1 = [*range(0,10+1)]
y1 = [-i*(i-14) for i in x1]
print(x1, y1)
```

```python
# create dataset 2
x2 = [*range(0,5+1)]
y2 = [3*i for i in x2]
print(x2, y2)
```

```python
# create dataset 3
x3 = [*range(5,10+1)]
y3 = [15+(i-5)**2 for i in x3]
print(x3, y3)
```

```python
import matplotlib.pyplot as plt
plt.plot(x1, y1, x2, y2, x3, y3)
plt.show()
```

```python
plt.plot(x1, y1, x2, y2, x3, y3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
plt.plot(x1, y1, label='dataset 1')
plt.plot(x2, y2, label='dataset 2')
plt.plot(x3, y3, label='dataset 3')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
