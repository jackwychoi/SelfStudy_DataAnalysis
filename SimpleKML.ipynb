```python
df = pd.read_csv('data/korea_gmonpro_20230103_152811_Sim1_45006.csv')
```

```python
import simplekml
```

```python
kml = simplekml.Kml()

for i in range(len(df.index)):
    kml.newpoint(name=df.iloc[i]['xNBID'], coords=[(df.iloc[0]['LON'], df.iloc[0]['LAT'])])

kml.save('data/kml_sample_a.kml')
```

```python
kml = simplekml.Kml()

for index, row in df.iterrows():
    pnt = kml.newpoint(name=row['xNBID'], coords=[(row['LON'], row['LAT'])])
    pnt.extendeddata.newdata('Index', index)
    for col in df.columns:
        pnt.extendeddata.newdata(col, row[col])
        
kml.save('data/kml_sample_b.kml')
```
