```python
premier_league_19_20['Season'] = '2019/20'
premier_league_19_20.head()
```

```python
premier_league_all = pd.concat([premier_league_15_16, 
                                premier_league_16_17, 
                                premier_league_17_18, 
                                premier_league_18_19, 
                                premier_league_19_20], axis=0, ignore_index=True)
```

```python
#excluding unecessary columns
premier_league_all.drop(['League', 'Table Type', 'KEY'], axis=1, inplace=True)
#renaming columns
premier_league_all.columns = ['position', 'team', 'games_played', 'wins', 'draws', 'losts', 'goals_for', 'goals_against', 'goals_difference', 'points', 'season']
#transforming some columns to 'int64' type
premier_league_all = premier_league_all.astype({'position': 'int64', 'wins': 'int64', 'draws': 'int64', 'losts': 'int64'})

premier_league_all.head(10)
```

```python
premier_league_all.groupby(by=['season', 'team']).mean().unstack()
```

```python
#function to get points in percentage
def percentage(val):
    return (val / (38*3)) * 100

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(26, 5))

#array with needed columns
itens = ['position', 'points', 'goals_for', 'goals_against']

premier_league_all_grouped = premier_league_all.groupby(by=['season', 'team']).mean().unstack()
plt.style.use('fivethirtyeight')
for i, item in enumerate(itens):
    if item != 'points':
        #other teams that played Premier League in all seasons (2015/16 to 2019/20), it explains the use of 'dropna()': ignore relegated teams
        premier_league_all_grouped.dropna(axis=1)[item].apply(percentage).plot(ax=ax[i], legend=False, color='grey', linewidth=1, alpha=0.4)
        
        #line for Klopp's Liverpool
        premier_league_all_grouped[item]['Liverpool'].apply(percentage).plot(ax=ax[i], legend=False, color='red', linewidth=7, alpha=0.8)
        
        #line for Guardiola's Manchester City (except season 2015/16, Manuel Pellegrini was in charge)
        premier_league_all_grouped[item]['Manchester City'].apply(percentage).plot(ax=ax[i], legend=False, color='lightblue', linewidth=7, alpha=1)

        ax[i].set_title(item, fontsize=14)
    else:
        #applying percentage function to get the percentage of points reached related to all possible
        premier_league_all_grouped.dropna(axis=1)[item].apply(percentage).plot(ax=ax[i], legend=False, color='grey', linewidth=1, alpha=0.4)
        premier_league_all_grouped[item]['Liverpool'].apply(percentage).plot(ax=ax[i], legend=False, color='red', linewidth=7, alpha=0.8, marker='o')
        premier_league_all_grouped[item]['Manchester City'].apply(percentage).plot(ax=ax[i], legend=False, color='lightblue', linewidth=7, alpha=1, marker='o')

        ax[i].set_title('percentage of possible points', fontsize=14)
        ax[i].set_ylim(0, 100)

    if item == 'position':
        ax[i].invert_yaxis()
        
    ax[i].xaxis.set_tick_params(labelsize=12)
    ax[i].set_xlabel('')
    ax[i].xaxis.grid(False)

plt.savefig('kloppguardiola.png', dpi=300)
plt.show()
```
