```python
import numpy as np
import pandas as pd
```

```python
melb = pd.read_csv("data/melb_data.csv", usecols = ["Address", "Regionname", "Type", "Rooms", "Distance", "Price"])
melb.head()
```

```python
melb["new1"] = melb.Price.where(melb.Price < 1400000, "class1")
melb.head()
```

```python
melb["new2"] = np.where(
   (melb.Rooms == 3) & (melb.Price > 1400000), 
   "class1", 
   "other"
)
melb.head()
```

```python
filters = [
   (melb.Rooms == 3) & (melb.Price > 1400000),
   (melb.Rooms == 2) & (melb.Price < 1400000),
   (melb.Price < 900000)
]
values = ["class1", "class2", "class3"]
```

```python
melb["new3"] = np.select(filters, values)
melb.head()
```

```python
melb["new4"] = np.select(filters, values, default="no_class")
melb.head()
```
