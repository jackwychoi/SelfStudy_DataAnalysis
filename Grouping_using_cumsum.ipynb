```python
import pandas as pd
```

```python
data = {'Col': ['cellNum', 'Param_1', 'Param_2', 'Param_3', 'cellNum', 'Param_1', 'Param_2', 'Param_3', 'cellNum', 'Param_1', 'Param_2', 'Param_3'],
        'Value': [0, 50, 50, 1, 3, 70, 20, 2, 11, 40, 30, 100]}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
df['Group'] = (df['Col'] == 'cellNum').cumsum()
```

```python
df
```

```python
result = df.pivot(index='Group', columns='Col', values='Value')
result
```

```python
result.index.name = None
result.columns.name = None
result
```
