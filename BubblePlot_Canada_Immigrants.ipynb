```python
import numpy as np  
import pandas as pd 
```

```python
df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
df.head(3)
```

```python
df.columns
```

```python
df = df.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName',]).set_index('OdName')
df.head(3)
```

```python
korea = df.loc['Republic of Korea']
japan = df.loc['Japan']
```

```python
k_normal = korea / korea.max()
j_normal = japan / japan.max()
```

```python
years = list(range(1980, 2014))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 8))
plt.scatter(years, korea, color='blue')
plt.scatter(years, japan, color='red')
plt.xlabel("Years", size=14)
plt.ylabel("Number of Immigrants", size=14)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(years, korea, 
                  color='navy', 
                 alpha=0.5,
                 s = k_normal * 2000)
plt.scatter(years, japan, 
                  color='purple', 
                 alpha=0.5,
                 s = j_normal * 2000,
                 )
plt.xlabel("Years", size=14)
plt.ylabel("Number of Immigrants", size=14)
```

```python
sort_k = sorted(korea)
sort_j = sorted(japan)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(years, korea, 
                  c=sort_k,
                 alpha=0.5,
                 s = k_normal * 2000)
plt.scatter(years, japan, 
                  c=sort_j,
                 alpha=0.5,
                 s = j_normal * 2000,
                 )
plt.xlabel("Years", size=14)
plt.ylabel("Number of Immigrants", size=14)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(years, korea, 
                  c=sort_k,
                 alpha=0.5,
                 s = k_normal * 2000)
plt.xlabel("Years", size=14)
plt.ylabel("Number of Immigrants of Korea", size=14)
```
