```python
import pandas as pd
import numpy as np

# Define the start time, end time, and frequency
start_time = '00:00:00'
end_time = '23:59:59'
freq = '15min'

# Create a time range between start and end times at the given frequency
time_range = pd.date_range(start='2023-01-19 ' + start_time, end='2023-02-18 ' + end_time, freq=freq)

# Create two random float value columns
values1 = np.random.uniform(low=-100, high=100, size=len(time_range))
values2 = np.random.uniform(low=-100, high=100, size=len(time_range))

# Create the DataFrame with the time series column and the two value columns
df = pd.DataFrame({'time': time_range, 'value1': values1, 'value2': values2})
df.head()
```

```python
import pandas as pd
import numpy as np

# Create a time range at 15 minute intervals for one day
time_range = pd.date_range(start='2023-01-19 00:00:00', end='2023-02-18 23:45:00', freq='15min')

# Create a dictionary to store the columns
columns = {'time': time_range}

# Create 98 additional columns with random integer values between 1 and 100
for i in range(1, 101):
    column_name = 'value' + str(i)
    values = np.random.randint(1, 10000, size=len(time_range))
    columns[column_name] = values

# Create a DataFrame with the time series column and the additional value columns
df = pd.DataFrame(columns)
df.head()
```

```python
import pandas as pd
import numpy as np

# Create a time range at 15 minute intervals for one day
time_range = pd.date_range(start='2023-01-19 00:00:00', end='2023-02-18 23:45:00', freq='15min')

# Create a dictionary to store the columns
columns = {'time': time_range}

# Create 98 additional columns with random integer values between 1 and 100
for i in range(1, 101):
    column_name = 'value' + str(i)
    values = np.random.choice([0, np.random.randint(1, 10000)], size=len(time_range), p=[0.98, 0.02])
    columns[column_name] = values

# Create a DataFrame with the time series column and the additional value columns
df = pd.DataFrame(columns)
df.head()
```
