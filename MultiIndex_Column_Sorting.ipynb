```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(np.random.randint(0,100000, size=(4, 8)), 
            index = pd.date_range('2016-01-01', periods=4),
            columns=pd.MultiIndex.from_tuples([('A', 'ENDC'), ('A', 'NRDC'), ('B', 'ENDC'), ('B', 'NRDC'),  ('C', 'ENDC'), ('C', 'NRDC'), ('D', 'ENDC'), ('D', 'NRDC')]))

df
```

```python
# calculate the column totals for the 'ENDC' columns
endc_totals = df.loc[:, pd.IndexSlice[:, 'ENDC']].sum()

# sort the DataFrame by the 'ENDC' column totals in descending order
df_sorted = df.loc[:, (endc_totals.sort_values(ascending=False).index)]

# create a new list of tuples to represent the reordered columns, including both 'ENDC' and 'NRDC'
new_cols = []
for col in df_sorted.columns:
    if col[1] == 'ENDC':
        new_cols.append(col)
        nrdc_col = (col[0], 'NRDC')
        if nrdc_col in df.columns:
            new_cols.append(nrdc_col)

# reorder the columns
df_sorted = df[new_cols]

df_sorted
```

```python
df_sorted.loc['Total'] = df_sorted.sum()
```
