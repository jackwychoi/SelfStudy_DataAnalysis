```python
import pandas as pd
import seaborn as sns 
```

```python
df = sns.load_dataset('taxis')
df.head()
```

## Grouping

```python
pd.crosstab([df.pickup_borough,df.pickup_zone],df.payment,\
            rownames = ["Borough","Zone"],\
            colnames = ["Payment Method"])
```

## Using aggfunc

```python
pd.crosstab(index = df.pickup_borough, columns = df.payment,\
                  values = df.distance, aggfunc = 'mean')
```

## Subtotal

```python
pd.crosstab(index = df.pickup_borough, columns = df.payment,\
                  values = df.total, aggfunc = 'sum',\
                  margins=True, margins_name="Total")
```

## Normalize the Crosstab

```python
pd.crosstab(index = df.pickup_borough, columns = df.payment,\
                  values = df.total, aggfunc = 'sum',\
                  normalize = "index")
```
