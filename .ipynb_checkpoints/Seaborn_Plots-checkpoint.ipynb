```python
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
data.shape
```

```python
sns.set_theme()
# For the image quality of the graphic. 
sns.set(rc={'figure.dpi':100})
# For the size of the graphics
sns.set(rc={'figure.figsize':(7,5)})
```

```python
sns.scatterplot(x='bill_length_mm', 
                y='bill_depth_mm', 
                data=data, 
                hue='species');
```

```python
sns.histplot(x='flipper_length_mm', data=data);
```

```python
sns.histplot(data=data, y='flipper_length_mm');
```

```python
sns.histplot(data=data, x='flipper_length_mm', binwidth=3);
```

```python
sns.histplot(data=data, x='flipper_length_mm', kde=True);
```

```python
sns.histplot(data=data, x='flipper_length_mm', hue='species');
```

```python
sns.barplot(data=data, x='species', y='flipper_length_mm');
```

```python
sns.barplot(data=data, x='species', y='flipper_length_mm', hue='sex');
```

```python
sns.boxplot(data=data, x='species', y='flipper_length_mm');
```

```python
sns.boxplot(data=data, x='species', y='flipper_length_mm', hue='sex');
```

```python
sns.violinplot(data=data, x='species', y='flipper_length_mm');
```

```python
sns.violinplot(data=data, x='species', y='flipper_length_mm', hue='sex');
```

```python
sns.FacetGrid(data, col='island', row='sex');
```

```python
sns.FacetGrid(data, col='island', row='sex').map(sns.histplot, 'flipper_length_mm');
```

```python
sns.FacetGrid(data, col='island', row='sex').map(sns.distplot, 'flipper_length_mm');
```

```python
sns.pairplot(data, hue='species', height=3);
```

```python
sns.pairplot(data, hue='species', diag_kind='hist');
```

```python
sns.heatmap(data.corr());
```

```python
sns.heatmap(data.corr(), annot=True);
```
