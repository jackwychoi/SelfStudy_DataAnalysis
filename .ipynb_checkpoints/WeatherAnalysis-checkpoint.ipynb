```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/weatherHistory.csv')
df.head()
```

```python
missing_values_count = df.isnull().sum()
# get the number of missing data points per column
missing_values_count
```

```python
df['Precip Type'].unique()
```

```python
df.dropna(how='any', inplace=True)
```

```python
df['Formatted Date'].head()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
```

```python
df['Formatted Date'].head()
```

```python
df.set_index("Formatted Date", inplace = True)
df.head()
```

```python
#resample time series data from hourly daily data into monthly data.
df_resampled = df.resample('M').mean()
df_resampled.head()
```

```python
app_temp = df_resampled['Apparent Temperature (C)']
hum = df_resampled['Humidity']
```

```python
plt.title('Analysis of Humidity and Temperature (2006 to 2016)', fontsize=25)
plt.plot(hum, label="Average Humidity")
plt.plot(app_temp, label="Average Apparent Temperature")
plt.xlabel("Years (2006 to 2016)", fontsize=13)
plt.legend(loc=(1.01,0.8));
```

```python
df_Jan = df_resampled[df_resampled.index.month==1]
hum1 = df_Jan['Humidity']
app_temp1 = df_Jan['Apparent Temperature (C)']
plt.figure(figsize=(10,5))
plt.title('For January (2006 to 2016)', fontsize=25)
plt.plot(hum1, label="Average Humidity", marker=".")
plt.plot(app_temp1, label="Average Apparent Temperature", marker=".")
plt.xlabel("Years (2006 to 2016)", fontsize=13)
plt.legend(loc=(1.01,0.8));
```
