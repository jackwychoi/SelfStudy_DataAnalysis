```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.DataFrame(np.random.rand(10, 4), columns=['A','B','C','D'])
data.head()
```

```python
data.plot.bar();
```

```python
data.plot.bar(stacked=True);
```

```python
data.plot.barh(stacked=True);
```

```python
sns.set_palette('magma')
data.plot.area();
```

```python
data.plot.area(stacked=False);
```

```python
data.diff()
```

```python
data.diff().plot.box(vert=False,
                     color={'medians':'lightblue',
                            'boxes':'blue',
                            'caps':'darkblue'});
```

```python
data2 = pd.DataFrame(np.random.rand(100, 1), columns=['value']).reset_index()
data2['value'].plot();
data2['value'].rolling(10).mean().plot();
```

```python
data.plot.kde(); #distribution plot
data.plot.scatter(x='A',y='B', #scatterplot x and y
                  c='C', #color of data points
                  s=data['C']*200); #size of data points
data.plot.hexbin(x='C',y='D', #hexbin x and y
                 gridsize=18); #hexagon dimensions
```
