```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = datasets.load_iris()
x, y = iris.data,iris.target
y_1 = np.array(['setosa' if i == 0 else 'versicolor' if i == 1 else 'virginica' for i in y])
pd_iris = pd.DataFrame(np.hstack((x, y_1.reshape(150,1))),
                       columns=['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)','class'])
pd_iris.head()
```

```python
pd_iris.dtypes
```

```python
for col in pd_iris.columns:
    if col != 'class':
        pd_iris[col] = pd_iris[col].astype('float64')
pd_iris.dtypes
```

```python
g = sns.pairplot(pd_iris)
g.fig.set_size_inches(10, 10)
sns.set(style='whitegrid', font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class')
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 10)
sns.set(style='whitegrid', font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette=sns.color_palette('colorblind', n_colors=len(pd_iris['class'].unique())))
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 10)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                palette=['#dc2624', '#2b4750', '#45a0a2'])
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 10)
sns.set(style='whitegrid',font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette=sns.color_palette('colorblind', n_colors=len(pd_iris['class'].unique())),
                 vars=['sepal length(cm)','sepal width(cm)'])
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 5)
sns.set(style='whitegrid',font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette=sns.color_palette('colorblind', n_colors=len(pd_iris['class'].unique())),
                 x_vars=['sepal length(cm)','sepal width(cm)'],
                 y_vars=['petal length(cm)','petal width(cm)'])
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 5)
sns.set(style='whitegrid',font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette=sns.color_palette('colorblind', n_colors=len(pd_iris['class'].unique())),
                 kind='reg')
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 10)
sns.set(style='whitegrid',font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette=sns.color_palette('colorblind', n_colors=len(pd_iris['class'].unique())),
                 diag_kind='hist')
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 10)
sns.set(style='whitegrid',font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette=sns.color_palette('colorblind', n_colors=len(pd_iris['class'].unique())),
                 diag_kind='kde')
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 10)
sns.set(style='whitegrid',font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette=sns.color_palette('colorblind', n_colors=len(pd_iris['class'].unique())),
                 diag_kind=None)
sns.set(style='whitegrid')
g.fig.set_size_inches(10, 10)
sns.set(style='whitegrid',font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette='Set1',
                 corner=True)
g.fig.set_size_inches(10, 10)
sns.set(font_scale=1.5)
```

```python
g = sns.pairplot(pd_iris, hue='class',
                 palette='Set1',
                 markers=['o','v','X'], #marker style
                 plot_kws=dict(s=50, edgecolor="r", linewidth=1), #maker size, edge and edge width
                 diag_kws=dict(shade=True))
g.fig.set_size_inches(10, 10)
sns.set(font_scale=1.5)
```
